### Goals

The goal of this project is to use NBA player statistics from the past several seasons to predict who will be on the all-NBA teams in the 2017-2018 season. We will attempt to answer this question by developing a machine learning classification model that will first learn the difference between an all-NBA player and a non all-NBA player using the data from past seasons. Then we will apply that model to the current season's statistics to test if the model can predict accurately who won the award. Every since the 1988-1989 season, all-NBA awards have been given to 15 players, split up into 3 teams of 5, composed of two forwards, two guards, and one center.

### Questions

One of the issues that pop up with this dataset right away is that there is a large discrepancy in the distribution of labels. Since a very small proportion of players actually win this award, it might be difficult to train the model to learn what a winner looks like after splitting the data into a training and a test set.

```python
# Problem 1
# from Lecture 5, Data Frame and Visualization
import pandas as pd

def get_nba_data(endpt, params, return_url=False):

    ## endpt: https://github.com/seemethere/nba_py/wiki/stats.nba.com-Endpoint-Documentation
    ## params: dictionary of parameters: i.e., {'LeagueID':'00'}
    
    from pandas import DataFrame
    from urllib.parse import urlencode
    import json
    
    useragent = "\"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9\""
    dataurl = "\"" + "http://stats.nba.com/stats/" + endpt + "?" + urlencode(params) + "\""
    
    # for debugging: just return the url
    if return_url:
        return(dataurl)
    
    jsonstr = !wget -q -O - --user-agent={useragent} {dataurl}
    
    data = json.loads(jsonstr[0])
    
    h = data['resultSets'][0]['headers']
    d = data['resultSets'][0]['rowSet']
    
    return(DataFrame(d, columns=h))
```
